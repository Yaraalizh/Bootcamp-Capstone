```python
#Uplode File
HR=pd.read_csv("HRData.csv")
HR.head()
```

```python
#Check the data and explore it
HR.info()
```

```python
#Check the the missing values
HR.isnull().sum()
```

```python
#Check the duplications
HR=HR.duplicated()
HR
```

```python
#Remove Columns
HR =HR.drop(columns=['Random Number','EmployeeCount','Over18'])
```

```python
#Create new column for Age Group
def categorize_age(age):
    if 18 <= age < 26:
        return '18-25'
    elif 26 <= age <= 30:
        return '26-30'
    elif 31 <= age <= 35:
        return '31-35'
    elif 36 <= age <= 40:
        return '36-40'
    elif 41 <= age <= 45:
        return '41-45'
    elif 46 <= age <= 50:
        return '46-50'
    elif 51 <= age <= 55:
        return '51-55'
    else:
        return '55+'
HR['Age Group'] = HR['Age'].apply(categorize_age)
```
